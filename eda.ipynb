```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/Lenovo/Downloads/dataset/dailyclimate.csv")
df.shape
df.info()
```

```python
df.describe()
```

```python
summary = df.describe()
mean_values = summary.loc['mean']
mean_values_cleaned = mean_values.drop('Unnamed: 0')
plt.figure(figsize=(10, 6))
mean_values_cleaned.plot(kind='bar', color='skyblue')
plt.title("Mean Values for Numeric Columns (After dropping 'Unnamed' column)")
plt.xlabel("Numeric Columns")
plt.ylabel("Mean Value")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

summary = df.describe()
mean_values = summary.loc['mean']
skewness_values = df.skew()

# Drop the 'Unnamed: 0' column from both mean_values and skewness_values
new_mean_values = mean_values.drop('Unnamed: 0')
new_skewness_values = skewness_values.drop('Unnamed: 0')

# Plot mean values
plt.figure(figsize=(10, 6))
new_mean_values.plot(kind='bar', color='skyblue', label='Mean')
plt.title("Mean Values and Skewness")
plt.xlabel("Numeric Columns")
plt.ylabel("Values")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Plot skewness values
ax = plt.gca()
ax2 = ax.twinx()
new_skewness_values.plot(kind='line', color='orange', label='Skewness', ax=ax2)
ax2.set_ylabel('Skewness')
ax.legend(loc='upper left')
ax2.legend(loc='upper right')

# Highlight skewness values exceeding 3
for i, value in enumerate(new_skewness_values):
    if abs(value) > 3:
        ax2.annotate(f'{value:.2f}', xy=(i, value), xytext=(i, value + 0.1),
                     arrowprops=dict(facecolor='red', arrowstyle='->'))

plt.show()
```

```python
import seaborn as sns

# Calculate correlation matrix
correlation_matrix = df.corr()

# Plot correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
import itertools

# Get column names from the mean values DataFrame
columns = new_mean_values.index

# Plot scatter plots for mean values
plt.figure(figsize=(12, 8))
for i, (col1, col2) in enumerate(itertools.combinations(columns, 2)):
    plt.subplot(3, 2, i + 1)
    plt.scatter(df[col1], df[col2], color='orange', alpha=0.5)
    plt.xlabel(col1 + ' (Mean)')
    plt.ylabel(col2 + ' (Mean)')
    plt.title(f'Scatter Plot: {col1} vs {col2} (Mean)')
plt.tight_layout()
plt.show()
```
